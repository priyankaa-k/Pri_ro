Pandas method-1

```python
#cond is to identify gov_id's having  multiple participant_id's
cond = df.groupby('AP_Gov_ID1').filter(lambda x: x['AP_Participant_ID'].nunique() > 1)['AP_Gov_ID1'].unique()
cond
```

```python
#show df with above cond and sort with gov_id
res_df = df[df['AP_Gov_ID1'].isin(cond)].sort_values(by='AP_Gov_ID1')
res_df
```

```python
#if you want to see govt_id and its different participant_id along with count of participant_id's
res_df2 = res_df.groupby(['AP_Gov_ID1', 'AP_Participant_ID']).size().reset_index(name='participant_count')

res_df2
```

```python
#if you want to see each gov_id has how many different unique participant_id use below code
# Count unique participant IDs for each AP_Gov_ID1
dup = df.groupby('AP_Gov_ID1')['AP_Participant_ID'].nunique().reset_index(name='Unique_Participant_Count')
```

```python
# Filter only those with duplicates
res_df3 = dup[dup['Unique_Participant_Count'] > 1].sort_values(by='AP_Gov_ID1')

res_df3
```

```python
import pandasql as psql

#if you want to see all the values with grouping govid and their different participant_id's
query = """
SELECT *
FROM df
WHERE AP_Gov_ID1 IN (
    SELECT  AP_Gov_ID1
    FROM df
    GROUP BY AP_Gov_ID1
    HAVING COUNT(DISTINCT AP_Participant_ID) > 1
)
order by AP_Gov_ID1,AP_Participant_ID
"""
result_df = psql.sqldf(query, locals())
result_df
```

```python
#if you want to see govt_id and its different participant_id along with count of participant_id's
query2 = """
SELECT AP_Gov_ID1, AP_Participant_ID, COUNT(AP_Participant_ID) as participant_count
FROM df
WHERE AP_Gov_ID1 IN (
    SELECT AP_Gov_ID1
    FROM df
    GROUP BY AP_Gov_ID1
    HAVING COUNT(DISTINCT AP_Participant_ID) > 1
)
GROUP BY AP_Gov_ID1, AP_Participant_ID
ORDER BY AP_Gov_ID1, participant_count DESC
"""

result_df2 = psql.sqldf(query2, locals())
result_df2
```

```python
#if you want to see each gov_id has how many different unique participant_id use below code
query3 = """
SELECT AP_Gov_ID1, COUNT( distinct AP_Participant_ID) as unique_participant_count
FROM df
WHERE AP_Gov_ID1 IN (
    SELECT AP_Gov_ID1
    FROM df
    GROUP BY AP_Gov_ID1
    HAVING COUNT(DISTINCT AP_Participant_ID) > 1
)
GROUP BY AP_Gov_ID1
ORDER BY AP_Gov_ID1
"""

result_df3 = psql.sqldf(query3, locals())
result_df3
```
