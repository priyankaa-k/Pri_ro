Pandas method-1

```python
#cond is to identify gov_id's having  multiple participant_id's
cond = df.groupby('gov_ID').filter(lambda x: x['part_ID'].nunique() > 1)['gov_ID'].unique()
cond
```

```python
#show df with above cond and sort with gov_id
res_df = df[df['gov_ID'].isin(cond)].sort_values(by='gov_ID')
res_df
```

```python
#if you want to see govt_id and its different participant_id along with count of participant_id's
res_df2 = res_df.groupby(['gov_ID', 'part_ID']).size().reset_index(name='participant_count')

res_df2
```

```python
#if you want to see each gov_id has how many different unique participant_id use below code
# Count unique participant IDs for each AP_Gov_ID1
dup = df.groupby('gov_ID')['part_ID'].nunique().reset_index(name='Unique_Participant_Count')
```

```python
# Filter only those with duplicates
res_df3 = dup[dup['Unique_Participant_Count'] > 1].sort_values(by='gov_ID')

res_df3
```

```python
import pandasql as psql

#if you want to see all the values with grouping govid and their different participant_id's
query = """
SELECT *
FROM df
WHERE gov_ID IN (
    SELECT  gov_ID
    FROM df
    GROUP BY gov_ID
    HAVING COUNT(DISTINCT part_ID) > 1
)
order by gov_ID,part_ID
"""
result_df = psql.sqldf(query, locals())
result_df
```

```python
#if you want to see govt_id and its different participant_id along with count of participant_id's
query2 = """
SELECT gov_ID, part_ID, COUNT(part_ID) as participant_count
FROM df
WHERE gov_ID IN (
    SELECT gov_ID
    FROM df
    GROUP BY gov_ID
    HAVING COUNT(DISTINCT part_ID) > 1
)
GROUP BY gov_ID, part_ID
ORDER BY gov_ID, participant_count DESC
"""

result_df2 = psql.sqldf(query2, locals())
result_df2
```

```python
#if you want to see each gov_id has how many different unique participant_id use below code
query3 = """
SELECT gov_ID, COUNT( distinct part_ID) as unique_participant_count
FROM df
WHERE gov_ID IN (
    SELECT gov_ID
    FROM df
    GROUP BY gov_ID
    HAVING COUNT(DISTINCT part_ID) > 1
)
GROUP BY gov_ID
ORDER BY gov_ID
"""

result_df3 = psql.sqldf(query3, locals())
result_df3
```
